```python
%matplotlib inline
import pandas as pd
import matplotlib.pylab as plt
import numpy as np
import seaborn as sns
```

```python
df_ratings = pd.read_csv('./data/ratings.csv')
df_movie = pd.read_csv('./data/movies.csv')
df = pd.merge(df_ratings, df_movie, left_on='movieId', right_on='movieId', how='left')
df.head()
```

```python
rating_matrix = df.pivot_table(index=['movieId'],columns=['userId'],values='rating')
rating_matrix = rating_matrix.fillna(0)
rating_matrix
```

```python
from sklearn.metrics.pairwise import cosine_similarity
movie_similarity = cosine_similarity(rating_matrix)
df_movie_similarity = pd.DataFrame(movie_similarity, index=rating_matrix.index, columns=list(rating_matrix.index))
df_movie_similarity
```

```python
user_input = 'Jumanji (1995)'
movie_id = df[df['title']==user_input]['movieId'].values[0]
df_movie_reccomend = df_movie_similarity[[movie_id]].sort_values(by=2, ascending=False).reset_index(drop=False)
df_movie_reccomend = pd.merge(df_movie_reccomend, df_movie)
df_movie_reccomend.head(15)
```
